```python
import numpy as np
```

```python
np.random.seed(1)
```

```python
import random
```

```python
#1000 random integers between 0 and 50
x=np.random.randint(0,50,1000)
```

```python
print(x)
```

```python
#positive correlation with some noice
y=x+np.random.normal(0,50,1000)
```

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
matplotlib.style.use('ggplot')
plt.scatter(x,y)
plt.show
```

```python
#negative correlation with some noice
y=x-np.random.normal(0,50,1000)
print(y)
```

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
matplotlib.style.use('ggplot')
plt.scatter(x,y)
plt.show
```

```python
import pandas as pd
df = pd.DataFrame({'a': np.random.randint(0, 50,1000)})
df['b'] = df['a'] + np.random.normal(0,10,1000)
df['c'] = 100-df['a'] + np.random.normal(0,5,1000)
df['d'] = np.random.randint(0,50,1000)
```

```python
pd.plotting.scatter_matrix(df,figsize=(6,6))
plt.show()
```

```python
plt.matshow(df.corr())
plt.xticks(range(len(df.columns)),df.columns)
plt.yticks(range(len(df.columns)),df.columns)
plt.colorbar()
plt.show()
```
